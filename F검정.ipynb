1. 라이브러리 설치

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import matplotlib.pyplot as plt
import seaborn as sns
```

2. 데이터 프레임 만들기

```python
csv_test = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(csv_test)
df.columns
df[['Q30_1', 'Q30_2', 'Q30_3', 'Q30_4', 'Q30_5', 'Q30_6']]
df1 = df[['Q30_1', 'Q30_2', 'Q30_3', 'Q30_4', 'Q30_5', 'Q30_6']]
df1_mean = np.mean(df1, axis=1)
data = {
    'Age_group': df['Age_group'],
    'mean': df1_mean
}
df1 = pd.DataFrame(data)
print(df1)
```

4. F 검정 수행 + 각 연령별 평균

```python
model = ols('mean ~ C(Age_group)', data=df1).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```

```python
group_means = df1.groupby('Age_group')['mean'].mean()
print("\n연령대별 평균:")
print(group_means)
```

5. 시각화

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Age_group', y='mean', data=df1)
plt.title('Means by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mean')
plt.show()
```
