1-1. ANOVA 시행

```python
df_mean = pd.DataFrame()
df1['Mean_Q1'] = df1[['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']].mean(axis=1)
df_mean = df1[['Age_group', 'Mean_Q1']]
df_mean
```

```python
grouped_data = [df_mean[df_mean['Age_group'] == group]['Mean_Q1'] for group in df['Age_group'].unique()]
grouped_data
```

```python
f_val, p_val = stats.f_oneway(*grouped_data)
print(f"ANOVA 결과: F-통계량 = {f_val}, p-값 = {p_val}")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
tukey = pairwise_tukeyhsd(endog=df_mean['Mean_Q1'],
                          groups=df_mean['Age_group'],
                          alpha=0.05)
print(tukey) #투키
```

2. 일원분산분석 시행

```python
from scipy.stats import f_oneway
```

```python
df = pd.DataFrame(csv_test)
columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
filtered_df = df[~df[columns_to_check].isin([9]).any(axis=1)]
filtered_df
filtered_df[['Age_group', 'Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']]
df1 = filtered_df[['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']]
df1 = np.mean(df1, axis=1)
```

```python
agedata = filtered_df[['Age_group']]
df1 = pd.concat([df1, agedata], axis=1)
df1.columns = ['mean', 'Age_group']
df1
```

```python
grouped_data = df1.groupby('Age_group')['mean'].apply(list)
print(grouped_data)
```

```python
f_statistic, p_value = f_oneway(*grouped_data)
print("일원분산분석(One-way ANOVA) 결과:")
print("F 통계량:", f_statistic)
print("p-value:", p_value)
```

3. 주요변수 기술 통계 - 독립변수

```python
import pandas as pd
from scipy import stats
```

```python
data = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(data)
df=df[['Age_group', 'DQ2_1', 'DQ6', 'DQ7']]
age_group_6 = df[df['Age_group'] == 6]
age_group_6
```

```python
def descriptive_statistics(df):
    stats_summary = pd.DataFrame(index=['mean', 'std', 'skewness', 'kurtosis'])

    stats_summary['DQ2_1'] = [
        df['DQ2_1'].mean(),
        df['DQ2_1'].std(),
        stats.skew(df['DQ2_1']),
        stats.kurtosis(df['DQ2_1'])
    ]

    stats_summary['DQ6'] = [
        df['DQ6'].mean(),
        df['DQ6'].std(),
        stats.skew(df['DQ6']),
        stats.kurtosis(df['DQ6'])
    ]

    stats_summary['DQ7'] = [
        df['DQ7'].mean(),
        df['DQ7'].std(),
        stats.skew(df['DQ7']),
        stats.kurtosis(df['DQ7'])
    ]

    return stats_summary
```

```python
stats_summary = descriptive_statistics(df)
print(stats_summary)
```
