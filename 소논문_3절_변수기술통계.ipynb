```python
import pandas as pd
import numpy as np
import scipy.stats as stats
csv_test = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(csv_test)
df
```

```python
columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
df = df[~df[columns_to_check].isin([9]).any(axis=1)]
df=df[['Age_group', 'DQ2_1', 'DQ6', 'DQ7']]
age_group_6 = df[df['Age_group'] == 6]
age_group_6
```

```python
def descriptive_statistics(age_group_6):
    stats_summary = pd.DataFrame(index=['mean', 'std', 'skewness', 'kurtosis'])

    stats_summary['DQ2_1'] = [
        df['DQ2_1'].mean(),
        df['DQ2_1'].std(),
        stats.skew(df['DQ2_1']),
        stats.kurtosis(df['DQ2_1'])
    ]

    stats_summary['DQ6'] = [
        df['DQ6'].mean(),
        df['DQ6'].std(),
        stats.skew(df['DQ6']),
        stats.kurtosis(df['DQ6'])
    ]

    stats_summary['DQ7'] = [
        df['DQ7'].mean(),
        df['DQ7'].std(),
        stats.skew(df['DQ7']),
        stats.kurtosis(df['DQ7'])
    ]

    return stats_summary
```

```python
stats_summary = descriptive_statistics(df)
print(stats_summary)
```

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis
```

```python
df = df = pd.DataFrame(csv_test)
columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
df = df[~df[columns_to_check].isin([9]).any(axis=1)]
age_group_6 = df[df['Age_group'] == 6]
age_group_6
```

```python
age_group_6 = age_group_6[['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']]
age_group_6
```

```python
age_group_6_mean = pd.DataFrame(age_group_6)
age_group_6_mean['Mean_Q1'] = age_group_6_mean.mean(axis=1)
#age_group_6_mean = age_group_6_mean.loc[:,['Mean_Q1']]
age_group_6_mean
```

```python
mean_val = age_group_6_mean['Mean_Q1'].mean()
std_dev = age_group_6_mean['Mean_Q1'].std()
skewness = skew(age_group_6_mean['Mean_Q1'])
kurt = kurtosis(age_group_6_mean['Mean_Q1'])

print(f"Mean: {mean_val}")
print(f"Standard Deviation: {std_dev}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
```
