```python
import pandas as pd
import numpy as np
import scipy.stats as stats
csv_test = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(csv_test)
df
```

```python
columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
filtered_df = df[~df[columns_to_check].isin([9]).any(axis=1)]
print(filtered_df)
```

```python
filtered_df.columns
filtered_df[['Age_group', 'Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']]
df1 = filtered_df[['Age_group', 'Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']]
df1
```

```python
df1.groupby(by='Age_group').mean().reset_index()
df1_1=df1.groupby(by='Age_group').mean().reset_index()
df1_1
```

```python
df_mean_q1 = pd.DataFrame()
df_mean_q1['Mean_Q1'] = df1_1[columns_to_check].mean(axis=1)
print(df_mean_q1)
```

```python
df_Age_group=df1_1[['Age_group']]
df_mean_q1 = pd.concat([df_Age_group, df_mean_q1], axis=1)
df_mean_q1
```

```python
import statsmodels.api as sm
import pandas as pd
```

```python
X = df_mean_q1['Age_group']  # 독립 변수: Age_group
y = df_mean_q1['Mean_Q1']       # 종속 변수: Mean_Q1
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
