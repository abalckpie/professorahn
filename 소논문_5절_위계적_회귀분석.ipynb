```python
import statsmodels.api as sm
import pandas as pd
from statsmodels.formula.api import ols
csv_test = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(csv_test)
df
```

```python
columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
df = df[~df[columns_to_check].isin([9]).any(axis=1)]
df=df[['Age_group', 'DQ2_1', 'DQ6', 'DQ7']]
age_group_6_1 = df[df['Age_group'] == 6]
age_group_6_1
```

```python
df = pd.DataFrame(csv_test)
columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
df = df[~df[columns_to_check].isin([9]).any(axis=1)]
age_group_6_2 = df[df['Age_group'] == 6]
age_group_6_2 = age_group_6_2[['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']]
age_group_6_2 = pd.DataFrame(age_group_6_2)
age_group_6_2['Mean_Q1'] = age_group_6_2.mean(axis=1)
age_group_6_2
```

```python
age_group_6_3 = columns_to_check = ['Q1_1', 'Q1_2', 'Q1_3', 'Q1_4', 'Q1_5', 'Q1_6', 'Q1_7', 'Q1_8', 'Q1_9', 'Q1_10']
df = df[~df[columns_to_check].isin([9]).any(axis=1)]
df=df[['Age_group', 'Birth']]
age_group_6_3 = df[df['Age_group'] == 6]
age_group_6_3['Year'] = age_group_6_3['Birth'].astype(str).str[:4]
age_group_6_3['Age'] = 2022 - age_group_6_3['Year'].astype(int)
age_group_6_3 = age_group_6_3[['Age']]
print(age_group_6_3)
```

```python
age_group_6 = pd.concat([age_group_6_1, age_group_6_2, age_group_6_3], axis=1)
age_group_6 = age_group_6[['Age','DQ2_1', 'DQ6', 'DQ7', 'Mean_Q1']]
age_group_6
```

```python
model1 = ols('Mean_Q1 ~ Age', data=age_group_6).fit()
print(model1.summary())
```

```python
model2 = ols('Mean_Q1 ~ Age + DQ2_1', data=age_group_6).fit()
print(model2.summary())
```

```python
model3 = ols('Mean_Q1 ~ Age + DQ2_1 + DQ6', data=age_group_6).fit()
print(model3.summary())
```

```python
model1 = ols('Mean_Q1 ~ + Age + DQ2_1 + DQ6 + DQ7', data=age_group_6).fit()
print(model1.summary())
```
