```python
import pandas as pd
import numpy as np
import scipy.stats as stats
!pip install pingouin
import pingouin as pg
csv_test = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(csv_test)
df
```

```python
df.columns
```

```python
df_6 = df[df['Age_group']==6]
df_6
```

```python
sex_counts = df_6['Sex'].value_counts()
print("Sex 칼럼의 값 개수:\n", sex_counts)
```

```python
DQ2_1_counts = df_6['DQ2_1'].value_counts()
print("\nDQ2_1 칼럼의 값 개수:\n", DQ2_1_counts)
DQ6_counts = df_6['DQ6'].value_counts()
print("\nDQ6 칼럼의 값 개수:\n", DQ6_counts)
```

```python
sex_counts = df_6['Sex'].value_counts()
sex_percent = df_6['Sex'].value_counts(normalize=True) * 100
sex_result = pd.DataFrame({'Counts': sex_counts, 'Percentage (%)': sex_percent})
print("Sex 칼럼의 값 개수와 비율:\n", sex_result)
```

```python
dq2_1_counts = df_6['DQ2_1'].value_counts()
dq2_1_percent = df_6['DQ2_1'].value_counts(normalize=True) * 100
dq2_1_result = pd.DataFrame({'Counts': dq2_1_counts, 'Percentage (%)': dq2_1_percent})
print("\nDQ2_1 칼럼의 값 개수와 비율:\n", dq2_1_result)
```

```python
dq6_counts = df_6['DQ6'].value_counts()
dq6_percent = df_6['DQ6'].value_counts(normalize=True) * 100
dq6_result = pd.DataFrame({'Counts': dq6_counts, 'Percentage (%)': dq6_percent})
print("\nDQ6 칼럼의 값 개수와 비율:\n", dq6_result)
```
