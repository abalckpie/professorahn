1. 데이터 불러오기

```python
import pandas as pd
import numpy as np
csv_test = pd.read_csv('/content/2022_2022년 지능정보사회 이용자 패널 조사_Data(가중치포함).csv')
df = pd.DataFrame(csv_test)
df
```

2. 필요한 데이터 추출

```python
df.columns
df[['Age_group', 'Q30_1', 'Q30_2', 'Q30_3', 'Q30_4', 'Q30_5', 'Q30_6']]
df1 = df[['Age_group', 'Q30_1', 'Q30_2', 'Q30_3', 'Q30_4', 'Q30_5', 'Q30_6']]
df1
```

3. 연령 그룹에 따른 문항별 평균값 구하기

```python
df1.groupby(by='Age_group').mean().reset_index()
df1_1=df1.groupby(by='Age_group').mean().reset_index()
df1_1
```

4. 연령 제거 후 그룹에 따른 전체 평균 구하기

```python
del df1_1['Age_group']
df1_1.mean(axis='columns')
```
